```python
import pandas as pd
import os
from typing import List, Dict

assessment_dir = "assessment/"
assessment_files = {"japanese.csv": "Japanese",
                    "asante_twi.csv": "Asante Twi",
                    "jinghpaw.csv": "Jinghpaw",
                    "south_azerbaijani.csv": "South Azerbaijani",}

def compute_tqs(judgement_counts):
    num_correct = judgement_counts["Correct"]
    num_minor = judgement_counts["Minor error"]
    num_major = judgement_counts["Major error"]
    num_critical = judgement_counts["Critical error"]
    return (3 * num_correct + 2 * num_minor + 1 * num_major + 0 * num_critical) / (3 * sum(judgement_counts.values())) * 100

def plot_judgements(assessment_files: Dict[str, str],
                    eval_col: str = "Fine-grained evaluation",
                    pie: bool = True,
                    latex: bool = True) -> None:
    """Plot pie charts for each language's evaluation results."""
    for f, lang in assessment_files.items():
        file_path = os.path.join(assessment_dir, f)
        df = pd.read_csv(file_path)
        evaluations_series = df.loc[:49, [eval_col]] # 50 rows
        evaluations_list = evaluations_series[eval_col].tolist()
        evaluations_list = [e if type(e) == str else "Correct" for e in evaluations_list]
        evaluations = []
        for eval in evaluations_list:
            try:
                evals = eval.split(", ")
            except Exception as e:
                evals = []
            evaluations += evals
        
        evaluations = ["Other" if "Other" in eval else eval for eval in evaluations]
        evaluations = ["Correct" if type(eval) == float else eval for eval in evaluations]
        judgement_counts = Counter(evaluations)
        # print(judgement_counts)
        print(compute_tqs(judgement_counts))
        
        if latex:
            tex = r"""\begin{figure}[ht]
    \centering
    \begin{tikzpicture}
        \pie[
            text=legend,
            radius=\radius, % Adjust size
            sum=auto,
            color={
                green!60,
                yellow!60,
                orange!60,
                red!60
            }
        ]{"""
            for label in ["Correct", "Minor error", "Major error", "Critical error"]:
                tex += f"{judgement_counts[label]}/{label}, "
            tex = tex[:-2]  # Remove the last comma and space
            tex += r"""};
    \end{tikzpicture}
    """
            tex += r"\caption{Judgement Distribution for " + lang + "}\n"
            tex += r"\label{fig:judgement_" + lang + "}"
            tex += r"\end{figure}"
            print(tex)
        else:
            colors = {"Minor error": "yellow",
                    "Major error": "orange",
                    "Critical error": "red",
                    "Correct": "green"}
            pie_colors = [colors.get(label, "black") for label in judgement_counts.keys()]
            
            plt.rcParams['axes.prop_cycle'] = plt.cycler(color=pie_colors)
            plt.figure(figsize=(5, 5))
            plt.pie(judgement_counts.values(), labels=judgement_counts.keys(), autopct='%1.1f%%', startangle=140)
            plt.title(f'Judgement Distribution for {lang}')
            plt.show()
```

```python
plot_judgements(assessment_files, eval_col="Error severity")
```

## tex table

```python
import pandas as pd
from collections import Counter

assessment_dir = "assessment"
assessment_files = {"japanese.csv": "Japanese",
                    "asante_twi.csv": "Asante Twi",
                    "jinghpaw.csv": "Jinghpaw",
                    "south_azerbaijani.csv": "South Azerbaijani",}

categories = ["Correct",
              "Wrong grammar",
              "Wrong punctuation",
              "Wrong spelling",
              "Wrong capitalization",
              "Inaccurately added information",
              "Inaccurately omitted information",
              "Mistranslation", 
              "Unnatural translation",
              "Untranslated text", 
              "Wrong register",
              "Other"]

def counter_to_latex_table(counts: dict[str, Counter],
                           caption: str = "Judgement Category Counts") -> str:
    table = []
    table.append(r"\begin{table}[ht]")
    table.append(r"\centering")
    table.append(r"\small")
    table.append(r"\setlength{\tabcolsep}{4pt}")
    table.append(r"\begin{tabular}{lrrrr}")
    table.append(r"\toprule")
    table.append(r"Category & \texttt{twi} & \texttt{jpn} & \texttt{kac} & \texttt{azb} \\")
    table.append("\\midrule")
    
    category_counter = {cat: Counter() for cat in categories}
    for lang, counter in counts.items():
        for cat, count in counter.items():
            category_counter[cat][lang] = count
    
    for label, lang_count in category_counter.items():
        row = f"{label} & "
        for lang in counts.keys():
            if lang == "South Azerbaijani":
                row += f"{lang_count[lang]} \\\\"
            else:
                row += f"{lang_count[lang]} & "
                
        table.append(row)
    table.append("\\bottomrule")
    table.append("\\end{tabular}")
    table.append(f"\\caption{{{caption}}}")
    table.append("\\end{table}")
    return "\n".join(table)

def make_table_judgements(assessment_files,
                          eval_col):
    counts = {}
    for f, lang in assessment_files.items():
        file_path = os.path.join(assessment_dir, f)
        df = pd.read_csv(file_path)
        evaluations_series = df.loc[:49, [eval_col]] # 50 rows
        evaluations_list = evaluations_series[eval_col].tolist()
        evaluations_list = [e if type(e) == str else "Correct" for e in evaluations_list]
        evaluations = []
        for eval in evaluations_list:
            try:
                evals = eval.split(", ")
            except Exception as e:
                evals = []
            evaluations += evals
        
        judgement_counts = Counter(evaluations)
        print(judgement_counts)
        counts.update({lang: judgement_counts})

    latex_code = counter_to_latex_table(counts, caption=f"Assessment statistics.")
    print(latex_code)

make_table_judgements(assessment_files,
                      eval_col="Fine-grained evaluation")
```

# String metrics

```python
import jiwer

def evaluate_metrics(assessment_files: List[str],
                     assessment_dir: str = "assessment/") -> None:
    """Plot pie charts for each language's evaluation results."""
    results = {}
    for f, lang in assessment_files.items():
        print(f)
        file_path = os.path.join(assessment_dir, f)
        df = pd.read_csv(file_path)
        print(df.columns)

        refs = df[f"{lang}"].tolist()[:50]
        if lang == "Asante Twi":
            lang = "Twi"
        hyps = df[f"Corrected {lang} translation"].tolist()[:50]
        for i, (ref, hyp) in enumerate(zip(refs, hyps)):
            if isinstance(hyp, float):
                hyps[i] = ref
        
        cer = jiwer.cer(refs, hyps)
        results.update({lang: cer})
    return results
```

```python
evaluate_metrics(assessment_files)
```
